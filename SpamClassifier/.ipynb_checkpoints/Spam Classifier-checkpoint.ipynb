```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
from scipy.io import loadmat
```

```python
data=loadmat("spamTrain.mat")
data
```

```python
X=data['X']
y=data['y']
```

# Make email feature list from Word list

```python
def featurelist(a):
    ############################## Opening Email or Spam ############

    email=open(a).read()

    ############################# Removing Non words ################

    import re
    email = email.lower()
    email = re.sub('[0-9]+', ' number',email)
    email = re.sub('<[^<>]+>',' ',email)
    email = re.sub('(http|https)://[^\s]*', 'httpaddr',email)
    email = re.sub('[^\s]+@[^\s]+', 'emailaddr',email)
    email = re.sub('[$]+', 'dollar',email)
    email = re.sub(r'[^\w\s]',' ',email)

    ############################ Importing Vocablist #################

    vocablist=open('vocab.txt').read()
    vocablist=re.sub('[0-9]+'," ",vocablist)

    ########################### Creating Feature List #################

    import nltk
    from nltk.stem.porter import PorterStemmer
    ps=PorterStemmer()
    features=[]
    for i in vocablist.split():
        i=ps.stem(i)
        if email.find(i)==-1:
            features.append(0)
        else:
            features.append(1)
    return features
```

# Constructing A spam Classifier

```python
from sklearn.svm import SVC
classifier=SVC(kernel='linear',probability=True)
classifier.fit(X,y.ravel())
```

```python
testdata=loadmat("spamTest.mat")
testdata
```

```python
X_test=testdata['Xtest']
y_test=testdata['ytest']
```

```python
y_pred=classifier.predict(X)
y_pred
```

```python
classifier.score(X_test,y_test)
```

```python
classifier.score(X,y)
```
