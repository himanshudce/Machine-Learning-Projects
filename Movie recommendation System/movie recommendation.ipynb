```python
import pandas as pd
import numpy as np
```

```python
ratings=pd.read_csv('ratings.csv',sep=",")
ratings.head()
```

```python
movies=pd.read_csv('movies.csv',sep=',')
movies.head()
```

```python
del movies['genres']
del ratings['timestamp']
```

```python
ratings=pd.merge(movies,ratings)
ratings.head()
```

### Pivote_table for movies and rating by each user

```python
movieRatings=ratings.pivot_table(index=['userId'],columns=['title'],values='rating')
movieRatings.head()
```

```python
starwarsRating=movieRatings['Star Wars: Episode IV - A New Hope (1977)']
starwarsRating[starWarsRating>0]
```

```python
similarMovies=movieRatings.corrwith(starwarsRating)
similarMovies=similarMovies.dropna()
df=pd.DataFrame(similarMovies)
df.head(10)
```

```python
similarMovies.sort_values(ascending=False).head(10)
```

```python
movieStats=ratings.groupby('title').agg({'rating':[np.size,np.mean]})
movieStats.sort_values([('rating','size')],ascending=False).head(20)
```

```python
popularMovies=movieStats['rating']['size']>=100
movieStats[popularMovies].sort_values([('rating','mean')],ascending=False).head(10)
```

```python
df=movieStats[popularMovies].join(pd.DataFrame(similarMovies,columns=['similarity']))
df.head()
```

```python
df.sort_values(['similarity'],ascending=False).head(10)
```
