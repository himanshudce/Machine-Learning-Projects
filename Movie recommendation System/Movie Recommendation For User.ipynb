```python
import pandas as pd
import numpy as np
```

```python
ratings=pd.read_csv('ratings.csv',sep=",")
ratings.head()
```

```python
movies=pd.read_csv('movies.csv',sep=',')
movies.head()
```

```python
del movies['genres']
del ratings['timestamp']
```

```python
ratings=pd.merge(movies,ratings)
ratings.head()
```

```python
movieRatings=ratings.pivot_table(index=['userId'],columns=['title'],values='rating')
movieRatings.head()
```

```python
corrMatrix=movieRatings.corr()
corrMatrix.head()
```

```python
corrMatrix=movieRatings.corr(method='pearson',min_periods=100)
corrMatrix.head()
```

```python
similarMovies=corrMatrix['Star Wars: Episode IV - A New Hope (1977)'][corrMatrix['Star Wars: Episode IV - A New Hope (1977)']>0]
similarMovies.sort_values(ascending=False).head(10)
```

## Predicting Movies for a random User

```python
d={'Star Wars: Episode VI - Return of the Jedi (1983)':5,'Dark Knight, The (2008)':4,'Jurassic Park (1993)':4}
userrating=pd.Series(d)
userrating
```

```python
simCandidates=pd.Series()
```

```python
for i in range(len(userrating)):
    print('Adding similar to ' + userrating.index[i] )
    sims=corrMatrix[userrating.index[i]].dropna()
    sims=sims.map(lambda x:x*userrating[i])
    simCandidates=simCandidates.append(sims)
```

```python
simCandidates.sort_values(inplace=True,ascending=False)
simCandidates.head()
```

```python
simCandidates=simCandidates.groupby(simCandidates.index).sum()
simCandidates.sort_values(inplace=True,ascending=False)
simCandidates.head(10)
```

```python
filtersimCandidates=simCandidates.drop(userrating.index)
filtersimCandidates.head(10)
```
